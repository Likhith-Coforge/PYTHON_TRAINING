```python
def count(row):
    tropical_count = 0
    fruity_count = 0
    if 'tropical' in row.description and 'fruity' in row.description:
        tropical_count += 1
        fruity_count += 1
    elif 'tropical' in row.description:
        tropical_count += 1
    elif 'fruity' in row.description:
        fruity_count += 1
    return tropical_count, fruity_count
```

```python
reviews.apply(count,axis='columns')
```
