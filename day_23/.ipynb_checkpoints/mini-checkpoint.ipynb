```python
all_nan_rows = json_data.dropna(subset=['Flights Operated','Passengers Carried','Target Passengers'],how='all')
```

```python
all_nan_rows
```

```python
invalid_month_rows = json_data[~json_data.Month.isin(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])]
```

```python
invalid_month_rows
```

```python
filtered_data = json_data.drop(invalid_month_rows.Month)
```

```python
filtered_data
```

```python
values = {'Flights Operated':flights_operated_mean,'Passengers Carried':passengers_carried_mean}
```

```python
final_data = filtered_data.fillna(value=values)
```

```python
final_data
```

```python
renamed_data = final_data.rename(columns={'Month':'month','Flights Operated':'flights_operated','Passengers Carried':'passengers_carried','targer_passengers':'target_passengers'})
```

```python
renamed_data.head(1)
```

```python
len(renamed_data)
```

```python
final_data = renamed_data.drop_duplicates(subset=['month'])
```

```python
csv_data = csv_data.drop_duplicates(subset=['Month'])
```

```python
csv_data = csv_data.rename(columns={'Month':'month','Flights Operated':'flights_operated','Passengers Carried':'passengers_carried','Target Passengers':'target_passengers'})
```

```python
csv_data
```

```python
csv_data['year'] = 2023
```

```python
csv_data
```

```python
final_data['year']=2024
```

```python
final_data
```

```python
combined_data = pd.concat([csv_data,final_data],ignore_index=True)
```

```python
combined_data
```

```python
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
combined_data['month'] = pd.Categorical(combined_data['month'],categories=month_order,ordered = True)
combined_data
```

```python
## Task-3: Data Analysis
```

```python
combined_data['pct_diff_target'] = ((combined_data['passengers_carried']-combined_data['target_passengers'])/combined_data['target_passengers'])*100
```

```python
combined_data.dtypes
```

```python
combined_data['flights_operated'] = combined_data['flights_operated'].astype(float)
```

```python
combined_data['passengers_carried'] = combined_data['passengers_carried'].astype(float)
```

```python
combined_data
```

```python
target_not_achievement = combined_data[combined_data.pct_diff_target < 0][['month','year','pct_diff_target']]
```

```python
total_passengers_carried = combined_data.groupby('year').passengers_carried.sum()
```

```python
total_flights_operated = combined_data.groupby('year').flights_operated.sum()
```

```python
total_flights_operated
```

```python
total_passengers_carried
```

```python
avg_pct_diff_target = combined_data.groupby('year').pct_diff_target.mean()
```

```python
avg_pct_diff_target
```

```python
best_performance = combined_data.groupby('year').pct_diff_target.max()
```

```python
best_performance[2024]
```

```python
type(best_performance)
```

```python
best = combined_data[combined_data.pct_diff_target.isin([best_performance[2023],best_performance[2024]])][['month','year']]
```

```python
worst_performance = combined_data.groupby('year').pct_diff_target.min()
```

```python
worst_performance
```

```python
worst = combined_data[combined_data.pct_diff_target.isin([worst_performance[2023],worst_performance[2024]])][['month','year']]
```
